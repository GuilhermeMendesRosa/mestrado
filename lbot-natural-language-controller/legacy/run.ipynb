```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Defina o nome do modelo e o token de autenticação
model_name = "GuilhermeMendesRosa/LBotMovementLanguageConverter"
hf_token = "TOKEN"

# Carregue o tokenizer e o modelo usando o token para autenticação
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=hf_token)
model = AutoModelForCausalLM.from_pretrained(model_name, use_auth_token=hf_token)

# Verifique se há GPU disponível, se não, use a CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Mova o modelo para o dispositivo correto (GPU ou CPU)
model.to(device)

# Exemplo de entrada para converter (substitua pela entrada desejada)
input_text = "Conerta para LBot Movement Language: Ande 500cm pra frente e 70cm pra direita, depois volte 50cm"

# Tokenize a entrada e gere os tensores para o modelo
inputs = tokenizer(input_text, return_tensors="pt").to(device)

# Gere a saída utilizando o modelo
outputs = model.generate(**inputs)

# Converta os tokens gerados de volta para string e exiba o resultado
resultado = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Resultado da conversão:", resultado)
```
